```python
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict
import random
```

```python
class GridWorld:
    def __init__(self, size=4):
        self.size = size
        self.actions = [0, 1, 2, 3] 
        self.terminal_states = [(0, 0), (size - 1, size - 1)]
        self.reset()

    def reset(self):
        while True:
            self.agent = (np.random.randint(self.size), np.random.randint(self.size))
            if self.agent not in self.terminal_states:
                break
        return self.agent

    def step(self, action):
        i, j = self.agent
        if action == 0: i = max(i - 1, 0)
        elif action == 1: j = min(j + 1, self.size - 1)
        elif action == 2: i = min(i + 1, self.size - 1)
        elif action == 3: j = max(j - 1, 0)
        
        new_pos = (i, j)
        self.agent = new_pos
        reward = 0 if new_pos in self.terminal_states else -1
        done = new_pos in self.terminal_states
        return new_pos, reward, done

env = GridWorld()
```

```python
def td_prediction(env, policy, alpha=0.1, gamma=1.0, episodes=500):
    value_table = defaultdict(float)

    for _ in range(episodes):
        state = env.reset()
        finished = False

        while not finished:
            action = policy(state)
            next_state, reward, finished = env.step(action)
            value_table[state] += alpha * (reward + gamma * value_table[next_state] - value_table[state])
            state = next_state

    return value_table
```

```python
def choose_action(Q, state, epsilon, num_actions):
    if random.random() < epsilon:
        return random.randint(0, num_actions - 1)
    return np.argmax(Q[state])

def sarsa_learning(env, alpha=0.1, gamma=1.0, epsilon=0.1, episodes=500):
    Q = defaultdict(lambda: np.zeros(len(env.actions)))

    for _ in range(episodes):
        state = env.reset()
        action = choose_action(Q, state, epsilon, len(env.actions))
        finished = False

        while not finished:
            next_state, reward, finished = env.step(action)
            next_action = choose_action(Q, next_state, epsilon, len(env.actions))
            Q[state][action] += alpha * (reward + gamma * Q[next_state][next_action] - Q[state][action])
            state, action = next_state, next_action

    return Q
```

```python
def q_learning(env, alpha=0.1, gamma=1.0, epsilon=0.1, episodes=500):
    Q = defaultdict(lambda: np.zeros(len(env.actions)))

    for _ in range(episodes):
        state = env.reset()
        finished = False

        while not finished:
            action = choose_action(Q, state, epsilon, len(env.actions))
            next_state, reward, finished = env.step(action)
            best_future = np.max(Q[next_state])
            Q[state][action] += alpha * (reward + gamma * best_future - Q[state][action])
            state = next_state

    return Q
```

```python
def get_state_values(Q):
    return {state: np.max(action_values) for state, action_values in Q.items()}

def show_value_heatmap(V, title="Value Function"):
    grid = np.zeros((4, 4))
    for i in range(4):
        for j in range(4):
            grid[i, j] = V.get((i, j), 0)

    plt.imshow(grid, cmap="coolwarm", interpolation="nearest")
    plt.colorbar()
    plt.title(title)
    plt.show()
```

```python
def random_policy(state):
    return random.choice(env.actions)

V_td = td_prediction(env, policy=random_policy)
Q_sarsa = sarsa_learning(env)
Q_q = q_learning(env)

V_sarsa = get_state_values(Q_sarsa)
V_q = get_state_values(Q_q)

show_value_heatmap(V_td, "TD(0) Prediction")
show_value_heatmap(V_sarsa, "SARSA Value Function")
show_value_heatmap(V_q, "Q-Learning Value Function")
```
